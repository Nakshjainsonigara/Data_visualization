```python
import numpy as np
import pandas as pd
import random as r

from bokeh.io  import output_file
from bokeh.layouts import layout
from bokeh.plotting import figure,show

from bokeh.models import ColumnDataSource,LabelSet
from bokeh.models import Div,RangeSlider,Spinner,Column
```

```python
from bokeh.io import output_notebook
output_notebook()
```

# Single and multiple line plot

```python
data=pd.DataFrame({
    "year":pd.date_range('13-04-2024',periods=10,freq='M'),
    "sales1":np.linspace(200,1000,10),
    "sales2":[100,150,600,200,1000,900,846,720,900,300]
})

p=figure(title="Salesplot",x_axis_label="Sales",y_axis_label="Month")

p.line(data['year'],data['sales1'],legend_label="sales trend",line_width=3)
p.line(data['year'],data['sales2'],legend_label="sales trend",line_width=3,color="red")

show(p)
```

```python
data = pd.DataFrame({
    "year":pd.date_range('13-04-2024',periods=5,freq='Y'),
    'Sales1': [11234,13389,19800,17903,18783] , 
    'Sales2': [21234,23389,19800,27903,28783] , 
    'Sales3': [31234,33389,39800,27903,38783] 
})

p=figure(title="multiple Salesplot",x_axis_label="Sales",y_axis_label="Month")

p.line(data['year'].dt.year,data['Sales2'],legend_label="sales 2",line_width=3)
p.vbar(x=data['year'].dt.year,top=data['Sales1'],legend_label="sales 1",width=0.5,bottom=0,color="red")
p.circle(data['year'].dt.year,data['Sales3'],legend_label="sales 3",line_width=3,color="black",size=10)

show(p)
```

```python
p=figure(title="Plot Parameters",
        sizing_mode="stretch_width",
        height=400,
        width=800,
        x_axis_label="Years",
        y_axis_label="sales")

p.circle(x=data['year'].dt.year,y=data['Sales1'],fill_color="gray",size=10)
p.line(x=data['year'].dt.year,y=data['Sales1'],line_width=1.5,color="black",alpha=0.7)

p.xaxis.axis_label='years'
p.xaxis.axis_line_width=1
p.xaxis.axis_line_color="red"

p.yaxis.axis_label='Sales'
p.yaxis.major_label_text_color="deeppink"
p.yaxis.major_label_text_alpha=0.5
p.yaxis.major_label_orientation="vertical"

show(p)
```

# Bubble plot

```python
df=pd.read_excel(r"C:\Users\jainn\Downloads\gapminder_full.xlsx")
df=df[df['year']==2007]
```

```python
p=figure(
    title="bubble plot",
    height=400,
    width=800,
    x_range=(df['gdp_cap'].min()-1000,df['gdp_cap'].max()+2500),
    y_range=(df['life exp'].min(),df['life exp'].max()+10)

)
p.circle(x=df['gdp_cap'],y=df['life exp'],radius=df['population']/500000,fill_color="purple",fill_alpha=0.6,line_color="black",line_width=2,line_alpha=0.8)

source = ColumnDataSource(data=df[df['population']>100000000])

labels = LabelSet(
    x='gdp_cap',
    y='life exp',
    text='country',
    level='glyph',
    x_offset=3,
    y_offset=5,
    source=source,
    text_color="black"
)

p.add_layout(labels)
show(p)
```

# Combining multiple plots 

```python
s1=figure(height=300,width=350,title="Plot1")
s1.line(x=data['year'].dt.year,y=data['Sales1'])

s2=figure(height=300,width=350,title="Plot2")
s2.line(x=data['year'].dt.year,y=data['Sales2'])


s3=figure(height=300,width=350,title="Plot3")
s3.line(x=data['year'].dt.year,y=data['Sales3'])

from bokeh.layouts import row,column
show(row(s1,s2,s3))
```

```python
show(column(s1,s2,s3))
```

```python
from bokeh.layouts import gridplot
grid=gridplot([[s1,None,None],
             [None,s2,None],
             [None,None,s3]],height=250,width=350)
show(grid)
```

# Widgets

```python
p=figure(y_range=(1,len(df['life exp'])),width=500,height=250)
points=p.circle(y=df['gdp_cap'],x=df['life exp'],size=7)

div = Div(
    text="""
    <p>Select circle size using this contral element:</p>
    """,
    width=200,
    height=30
    )

spinner = Spinner(
    title="circle size",
    low=0,
    high=60,
    step=5,
    value=points.glyph.size,
    width=200
)
spinner.js_link("value",points.glyph,"size")

range_slider_x = RangeSlider(
    title="X Range size",
    start=df['life exp'].min(),
    end=df['life exp'].max(),
    step=1,
    value=(df['life exp'].min(), df['life exp'].max())
)
range_slider_x.js_link("value", p.x_range, "start", attr_selector=0)
range_slider_x.js_link("value", p.x_range, "end", attr_selector=1)

range_slider_y = RangeSlider(
    title="Y Range size",
    start=df['gdp_cap'].min(),
    end=df['gdp_cap'].max(),
    step=1,
    value=(df['gdp_cap'].min(), df['gdp_cap'].max())
)
range_slider_y.js_link("value", p.y_range, "start", attr_selector=0)
range_slider_y.js_link("value", p.y_range, "end", attr_selector=1)

from bokeh.layouts import layout
layout = layout([[div, spinner], [range_slider_x, range_slider_y], [p]])
show(layout)
```
